```python
from mylibrary.connections import cursor, conn
import pandas as pd 

sql = """
select icscode as icscode_depart, station_name as station_name_depart from tt_h.stations_maps_data
where natrail_journey_minutes_cw is null or natrail_journey_minutes_pf is null
"""

df1 = pd.read_sql(sql, conn)
df1["join"] = 1
```

```python
# Get list of all cl stations 
sql = """
SELECT station_name as station_name_arrive, icscode as icscode_arrive, "1516_entries_&_exits"
from tt_h.all_stations as s, tt_h.inner_london as ie
where ST_Contains(ie.geom, s.geom)
order by cast("1516_entries_&_exits" as float) desc
limit 10
"""

df2 = pd.read_sql(sql, conn)
df2["join"] = 1
df2
```

```python
df_all = df1.merge(df2, on="join")
```

```python
from mylibrary.secrets import app_id, app_key
from mylibrary.tfl_helpers import get_journeyplanner_raw_content, status_of_tfl_response_journey, get_journeyplanner_results
counter = 1
for r in df_all.iterrows():
    counter +=1 
    print(counter)
    index = r[0]
    row = r[1]
    
    param_dict = [{
        "from": row["icscode_depart"],
        "to": row["icscode_arrive"],
        "id": app_id,
        "key": app_key,
        "time": "0900",
        "date": "20170906",
        "nationalsearch": True}]
    
    try:
        journey_object = get_journeyplanner_results(param_dict)

        d = 100000
        this_journey = None
        for j in journey_object['journey']['journeys']:
            if j['duration'] < d:
                d = j['duration']
                this_journey = j

        df_all.loc[index, "duration"] = j['duration']
        df_all.loc[index, "arrival_icscode"] = j['legs'][-1]['arrivalPoint']['icsCode']
    except:
        print("problem with {} {}".format(row["station_name_depart"], row["station_name_arrive"]))
    
    
    

df_all

```

```python
df_all = df_all[df_all["station_name_arrive"] != "Stratford"]
```

```python
df_all_2 = df_all.sort_values(["station_name_depart", "duration"]).drop_duplicates("station_name_depart")
```

```python
df_all_3 = df_all_2[pd.notnull(df_all_2["arrival_icscode"])]
```

```python
df_all_4 = df_all_3[["icscode_depart","station_name_depart","station_name_arrive","arrival_icscode"]]
```
